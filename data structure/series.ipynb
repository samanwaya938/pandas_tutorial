### Without index

```python
import pandas as pd

a = pd.Series([10,20,30])
print(a)
```

```python
print(a[0])
```

### With index

```python
b = pd.Series([10,20,30], index=["A","B","C"])
print(b)
```

```python
print(b["A"])
```

```python
marks = pd.Series([85,90,88], index=["Math", "English", "Science"])
print(marks)
```

```python
print(marks["Science"])
```
