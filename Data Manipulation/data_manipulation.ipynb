```python
import pandas as pd

df = pd.read_csv("titanic.csv")
df.head()
```

### Adding Constant column

```python
df["Country"] = "UK"
df.head()
```

### Calculated column

```python
df["FamilySize"] = df["SibSp"] + df["Parch"]
df.head()
```

### New column based on Condition

```python
df["Is_child"] = df["Age"] < 18
df
```

### Update Values

```python
df.loc[0, "Age"] = 25
df.head()
```

### Upadte values using condition

```python
df.loc[df["Age"] < 18, "Category"] = "Child"
df.head()
```

```python
df.loc[df["Age"] >= 18, "Category"] = "Adult"
df.head()
```

```python
import numpy as np 

df["Cat"] = np.where(df["Age"] < 18, "Child", "Adult")
df.head()
```

### .apply function

```python
df["Cat2"] = df["Age"].apply(lambda x: "Child" if x < 18 else "Adult")
df.head()
```

```python
import pandas as pd

df1 = df.iloc[:5]
df2 = df.iloc[5:10]

new_df = pd.concat([df1, df2])
new_df
```
