```python
import pandas as pd

data = {
  "Name": ["A","A","B","B"],
  "Subject": ["Math", "Science", "Math", "Science"],
  "Marks": [90,80,85,88]
}

df = pd.DataFrame(data)
df
```

### Pivot

```python
pivot_df = df.pivot(index="Name", columns="Subject", values="Marks")
pivot_df
```

### Melt

```python
df_melt = pd.melt(pivot_df.reset_index(), id_vars=["Name"], value_vars=["Math", "Science"])
df_melt
```

### Stack

```python
df_stack = pivot_df.stack()
df_stack
```

### Unstack

```python
df_unstack = df_stack.unstack()
df_unstack
```
