```python
import pandas as pd

data = {
  "Name":["A","B","C","D","E","F","G","H"],
  "Marks": [85,75,80,60,63,73,81,90]
}

df = pd.DataFrame(data)
print(df)
```

### .head function

```python
print(df.head(8))
```

### .tail function

```python
print(df.tail(7))
```

### .shape function

```python
print(df.shape)
```

### .info function

```python
print(df.info())
```

### .describe function

```python
print(df.describe())
```

### .column and .index

```python
print(df.columns)
print(df.index)
```

### value_counts function

```python
data = {
  "Name":["A","B","A","D","C","F","D","H"],
  "Marks": [85,75,80,60,63,73,81,90]
}

df2 = pd.DataFrame(data)
print(df2)
```

```python
print(df2["Name"].value_counts())
```
