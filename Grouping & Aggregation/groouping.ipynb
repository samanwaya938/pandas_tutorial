```python
import pandas as pd

df = pd.read_csv("titanic.csv")
df.head()
```

```python
a = df.groupby("Sex")["Age"].mean()
a
```

```python
b = df.groupby("Pclass")["PassengerId"].count()
b
```

```python
c = df.groupby("Pclass")["Fare"].agg(["mean", "max", "min"])
c
```

```python
d = df.groupby(["Sex", "Pclass"])["Fare"].mean()
d
```

```python
e = df.groupby(["Sex", "Pclass"])["Fare"].mean().reset_index()
e
```

```python
f = df.groupby("Pclass")["Fare"].mean().sort_values(ascending=True)
f
```

```python
g = df.groupby("Sex")["Fare"].mean().sort_values(ascending=True)
g
```
