```python
import pandas as pd

df = pd.read_csv("titanic.csv")
df.head()
```

### Vertical Concatination

```python
df1 = df.iloc[:5]
df2 = df.iloc[5:10]

new_df = pd.concat([df1,df2])
new_df
```

### Horizontal Concatination

```python
new_df2 = pd.concat([df1,df2], axis=1)
new_df2
```

### Merge

```python
extra = pd.DataFrame({
  "PassengerId": [1,2,3],
  "VIP": ["Yes","No", "Yes"]
})

extra
```

#### Inner merge (Only matching rows)

```python
inner_merged_df = pd.merge(df, extra, on= "PassengerId" )
inner_merged_df
```

#### Left join (All left data + matching right other data)

```python
left_merged_df = pd.merge(df, extra, on="PassengerId", how="left")
left_merged_df.head(10)
```

#### Outer Join (All data from both)

```python
outer_merged_df = pd.merge(df, extra, on="PassengerId", how="outer")
outer_merged_df
```

#### Right join ( All data from right + matcing from left)

```python
right_merged_df = pd.merge(df, extra, on="PassengerId", how="right")
right_merged_df.head(10)
```

### Join

```python
df1 = df.set_index("PassengerId")
extra = extra.set_index("PassengerId")

join_df = df1.join(extra)

join_df
```
